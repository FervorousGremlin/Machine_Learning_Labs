```python
import numpy as np

def load_data(file_name):
    return np.genfromtxt(file_name, delimiter=',', skip_header=1, usecols=(1, 2))

train_data = load_data('lab_1_train.csv')
test_data = load_data('lab_1_test.csv')

X_train, y_train = train_data[:, 0], train_data[:, 1]
X_test, y_test = test_data[:, 0], test_data[:, 1]
```

```python
W = np.random.rand()
b = np.random.rand()

learning_rate = 0.001
epochs = 100000
threshold = 1e-6 

for epoch in range(epochs):
    y_pred = W * X_train + b
    error = y_train - y_pred
    cost = np.mean(error ** 2)
    dW = -2 * np.mean(error * X_train)
    db = -2 * np.mean(error)
    W -= learning_rate * dW
    b -= learning_rate * db
    
    if np.abs(dW) < threshold and np.abs(db) < threshold:
        print(f"Stopping early at epoch {epoch}")
        break
    
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, W: {W}, b: {b}, Cost: {cost}")
```

```python
y_pred_test = W * X_test + b
test_cost = np.mean((y_pred_test - y_test) ** 2)
print(f"Test Cost: {test_cost}")
```

```python
import matplotlib.pyplot as plt
x_min = min(np.min(X_train), np.min(X_test))
x_max = max(np.max(X_train), np.max(X_test))

x_grid = np.linspace(x_min, x_max, 1000)

y_grid = W * x_grid + b


plt.scatter(X_train, y_train, color='blue', label='Train Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(x_grid, y_grid, color='red', label='Regression Line')  
plt.legend()
plt.show()
```
